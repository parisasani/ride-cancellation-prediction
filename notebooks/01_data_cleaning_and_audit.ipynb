```python
import pandas as pd

df = pd.read_csv('uber.csv')

print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

```python
print(df['Booking Status'].value_counts())
```

```python
import numpy as np

cancel_status = [
    'Cancelled by Driver',
    'Cancelled by Customer',
    'No Driver Found'
]

df['target_cancelled'] = np.where(
    df['Booking Status'].isin(cancel_status), 1,
    np.where(df['Booking Status'] == 'Completed', 0, np.nan)
)
```

```python
print(df['target_cancelled'].value_counts(dropna=False))
```

```python
df_step1 = df[df['target_cancelled'].notna()].copy()
```

```python
print(df_step1.shape)
print(df_step1['target_cancelled'].value_counts())
```

```python
drop_cols = [
    'Booking Status',
    'Cancelled Rides by Customer',
    'Reason for cancelling by Customer',
    'Cancelled Rides by Driver',
    'Driver Cancellation Reason',
    'Incomplete Rides',
    'Incomplete Rides Reason',
    'Booking ID',
    'Customer ID'
]

df_step1_clean = df_step1.drop(columns=drop_cols, errors='ignore')
```

```python
print(df_step1_clean.columns)
print(df_step1_clean.shape)
```
